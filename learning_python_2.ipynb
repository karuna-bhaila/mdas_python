```python
# Example 3
def print_grade(item):
    if item >= 90:
        print(f'{item}: A')
    elif item >= 80:
        print(f'{item}: B')
    elif item >= 70:
        print(f'{item}: C')
    else:
        print(f'{item}: F')
    
print_grade(75)
print_grade(88)

print()

# Function definition with a return values
def get_grade(item):
    if item >= 90:
        return 'A'
    elif item >= 80:
        return 'B'
    elif item >= 70:
        return 'C'
    
    return 'F'

print(get_grade(75))
print(get_grade(88))

x = get_grade(78)
```

```python
# Example 3
def roundNum(num):
    remainder = num % 1
    if remainder < 0.5:
        return num - remainder
    else:
        return num + (1-remainder)
    
print(roundNum(2.4))
print(roundNum(5.8))
```

```python
# Global vs Local variables

def test(x):
    y = x + z
    print(y)
    
z = 4
test(3)
print(y)
```

```python
# Example 4: Generic scale function
# Scales from source range to destination range
def scale(val, src, dst=[-1,1]):
    return (int(val-src[0]) / (src[1]-src[0]) * (dst[1]-dst[0]) + dst[0])

print(scale(49, (0,100), (-50,50)))
print(scale(49, (-100,100), (-50,50)))
print(scale(49, (-100,100)))
```

```python
# Create ndarray
x = np.array([1,2,3,4,5,6])
print('x:\n',x)

y = np.array([[1,2,3],[4,5,6]])
print('y:\n',y)

z = np.ones_like(x) #np.zeros_like
print('z:\n',z)

p = np.zeros([6]) #np.ones
print('p:\n',p)

q = np.full([3,4,], fill_value=5)
print('q:\n',q)

s = np.arange(1, 10, 2)
print('s:\n',s)

r = np.random.randint(1,7, size=[5,5])
print('r:\n',r)

# the different dimensions in an ndarray are called axes
# In a 2darray, rows are on axis 0 and columns are on axis 1
```

```python
# Some attributes
print('shape:',r.shape)
print('datatype:',r.dtype)
print('No. of dimensions:',r.ndim) 
print('Max:',r.max())
print('Min:',r.min())
print('Mean:',r.mean())
```

```python
# ndarray operations
# transpose
r = np.random.randint(1,10, size=[4,4])

print(r)
print(r.T)

print()

# reshape
print(r.reshape([2,8])) #2,2,4

print()

#ravel
print(r.ravel())
print(r.flatten())
```

```python
# Indexing and slicing ndarrays
r = np.zeros([4,4])
print(r)

print()

r[0] = [1,2,3,4]
print(r)

print()

r[1:,0] = [2,3,4]
print(r)

print()

r[1:,1:] = np.random.randint(1, 4, size=[3,3])
print(r)

print()

np.fill_diagonal(r, 0)
print(r)
```

```python
x = np.random.randint(1,100, size=[7])
print(x)

# Indexing with arrays of indices

indx_arr = [0,2,3,6]
print(x[indx_arr])

# Logical indexing
# Indexing with Boolean arrays
# Use a boolean masking array instead of indices
# size of boolean array must be same as the array

mask = [True,False,True,True,False,False,True]
print(x[mask])
```

```python
# Broadcasting
# With a scalar

a = np.random.randint(1,100, size=[3,5])
print('a:\n',a)

b = a + 1
print('b:\n',b)

c = b + [1,-1,0,-1,1]
print('c:\n',c)
```

```python
# Broadcasting with another ndarray
# element-wise operation
x = np.random.randint(1,9, size=[2,3])
y = np.random.randint(1,9, size=[2,3])

z = x * y  # element wise 

print('x:\n',x)
print('y:\n',y)
print('z:\n',z)
```

```python
# Matrix mulitplication
# x dot y: no. of columns in x = no. of rows in y
z = x.dot(y.T)
print('z:\n',z)
```

```python
# Output even numbers in an ndarray
r = np.random.randint(1,100, size=[5,5])
print('r:\n',r)

# for i in range(r.shape[0]):
#     for j in range(r.shape[1]):
#         if r[i][j]%2 == 0:
#             print(r[i][j])
            
for elem in r.flatten():
    if elem%2 == 0:
        print(elem)
```

```python
# Vectorization

# Output even numbers in an ndarray with broadcasting and logical indexing

mask = r%2==0
print('mask:\n',mask)
print(r[mask])

print()

print(r[r%2==0])

# Before using a for loop on a numpy array, check if it can be vectorized first
# The vectorized method is more computationally efficient and preferred over loops
```

# Min-max normalization
Scale values between [0,1] 
## $\frac{X - X_{min}}{X_{max} - X_{min}}$

```python
display(iris.columns)

for col in iris.columns[0:4]:
    iris[col] = (iris[col] - iris[col].min()) / (iris[col].max() - iris[col].min())
    
display(iris)

display(iris.describe())
```

# Standardization
Scale such that values are centered around the mean with a unit standard deviation
## $\frac{X - \mu}{\sigma}; \mu = \text{Mean of } X, \sigma = \text{Std. of X}$

```python
for col in iris.columns[0:4]:
    iris[col] = (iris[col] - iris[col].mean()) / iris[col].std()
    
display(iris)

display(iris.describe())
```

# Discretization
Transform continuous variables into a discrete form

```python
iris['PetalLength'] >= 0
```

```python
for col in iris.columns[0:4]:
    iris[col] = (iris[col] >= 0).astype(int)
display(iris.describe())
```
